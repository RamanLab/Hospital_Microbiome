```python
# Import necessary libraries
import numpy as np
import pandas as pd

# Set the path for the data
input_path = '/mnt/Local_Disk_1/2_Hospital_Microbiome/Data/Output_data/'
```

```python
# Finding pathogen
# Loading pathogen data
pathogen_data = pd.read_csv(input_path + 'gcPathogen_20241019.txt', sep = '|')

pathogen_data.head()
```

```python
pathogen_data.columns
```

```python
# Loading the species data
taxa_data = pd.read_csv(input_path + 'taxa_binary.csv', sep = ',', index_col=0)

taxa_data.head()
taxa_data['Present'] = taxa_data.index.to_series().apply(lambda x: 'Yes' if any(pathogen_data['Pathogen name'] == x) else 'No')
taxa_data['Biosafety level'] = taxa_data.index.to_series().apply(lambda x: pathogen_data[pathogen_data['Pathogen name'] == x]['Biosafety level'].values[0] if any(pathogen_data['Pathogen name'] == x) else None)
taxa_data['Biosafety level'].replace(np.nan, 'NA', inplace=True)
taxa_data['Biosafety level'].fillna('NA', inplace=True)
taxa_data
```

```python
# Save the data
taxa_data.to_csv(input_path + 'taxa_binary_pathogen.tsv', sep = '\t')
```
